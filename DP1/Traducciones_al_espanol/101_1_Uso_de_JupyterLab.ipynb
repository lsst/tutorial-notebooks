#### 2.1.1. Imagen

La opcin de imagen del servidor corresponde a la versin del entorno de software (Pipelines Cientficas de LSST - LSST Science Pipelines) que se usar en la sesin de JupyterLab.

**La imagen recomendada** ha sido seleccionada por el personal de Rubin como una versin estable, y la documentacin y tutoriales de Rubin siempre se mantienen actualizados con la imagen recomendada.

La imagen recomendada se actualiza en escalas de semanas a meses, de manera que los cambios son incrementales y es ms fcil para toda la comunidad evolucionar y mantener el cdigo y la documentacin.

Elegir siempre la imagen recomendada, a menos que se sepa que es necesaria una versin anterior o ms reciente de las Pipelines Cientficas de LSST (por ejemplo, si est reejecutando cdigo personalizado antiguo o realizando trabajo de desarrollo).

**Las imgenes antiguas** estn disponibles en el men desplegable, pero no se conservan indefinidamente.
Se recomienda actualizar el cdigo personalizado para que funcione con la ltima imagen recomendada.

#### 2.1.2. Opciones

Las opciones corresponden al tamao del servidor en trminos de CPUs (Central Processing Units - unidades centrales de procesamiento) y RAM (Random Access Memory - memoria de acceso aleatorio, en gigabytes).

Seguir las siguientes indicaciones como gua general.

**Large** (grande): para recuperar y visualizar datos, ejecutar notebooks de tutoriales.

**Small** (pequeo): para simplemente navegar o editar archivos sin acceder a datos.

**Enable debug logs** (habilitar registros de depuracin): no marcar esta casilla. El depurador [Debugger](https://jupyterlab.readthedocs.io/en/latest/user/debugger.html) no est habilitado en la Plataforma Cientfica de Rubin.

**Reset user environment** (restablecer entorno): marcar esta casilla para restaurar los archivos de configuracin del *software* a los valores predeterminados (elimina *software* instalado por la persona usuaria).

## 7. Apagar el servidor

Se recomienda siempre guardar y cerrar todos los archivos, y elegir el men de archivo File y luego la opcin para guardar todo, salir y cerrar sesin Save all, exit, and log out al final de cada sesin.

Esto es importante para preservar recursos para otras personas y para garantizar que cada vez que se ingrese nuevamente al RSP se haga en un estado conocido.

Para ayudar a evitar problemas con instancias obsoletas, las sesiones se cerrarn automticamente despus de 5 das de inactividad, o despus de 25 das.
